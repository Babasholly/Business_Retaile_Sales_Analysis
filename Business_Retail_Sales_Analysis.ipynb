### Using AI generate data dictionary

- Month : The month in which the data is recorded
- Year : The year in which the data is recorded
- Total Orders: The total number of orders placed during the specified month and year.
- Gross Sales: The total sales revenue generated before any deductions such as discounts or returns.
- Discounts: The amount deducted from the gross sales due to discounts offered to customers.
- Returns: The amount deducted from the gross sales due to returned items.
- Net Sales: The total sales revenue after deducting discounts and returns from the gross sales.
- Shipping : The amount charged for shipping products to customers.
- Total Sales: The total revenue generated from sales after deducting discounts, returns, and adding shipping charges.

#### Data Exploration via Visualization

```python
cat_columns
```

```python
plt.figure(figsize=(30,8))
sb.countplot(x='Total Sales', data=brs_df)
plt.title('Countplot of Total Sale')
```

```python
num = brs_df.select_dtypes(include=['float', 'int'])
num_columns = [i for i in num.columns]
```

```python
num_columns
```

```python
for column in num_columns:    
    plt.hist(brs_df[column], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Gross Sales')
    plt.show()
```

```python
for column in num_columns:
    plt.figure(figsize=(20,8))
    plt.boxplot(brs_df[column], notch=True, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'), medianprops=dict(color='red', linewidth=2), whiskerprops=dict(color='green', linewidth=1.5, linestyle='--'), capprops=dict(color='purple', linewidth=1.5), flierprops=dict(marker='o', markersize=8, markerfacecolor='yellow', markeredgecolor='black'), vert=False)
    plt.title(f'Box plot of {column}')
    plt.ylabel(column)
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.tight_layout() 
    plt.show()
```

```python
for column in num_columns:
    print(f'Stats for {column}')
    print(f'column : {brs_df[column].sum()}')
    print(f' Average : {brs_df[column].mean()}')
    print('-------------------------')
```

###### From the statistics presented above, it could be observed that the average of discounts is low when compared to the average of total sales . We can therefore conclude that a higher discount will encourage higher patronage for the business thereby increase their total sale

```python
plt.figure(figsize=(15,8))
sb.lineplot(data=brs_df, x='Year', y='Gross Sales')
plt.title('Trend of Gross Sales Over Year')
plt.xlabel('Year')
plt.ylabel('Gross Sales')
plt.show()
```

from the graph above the gross sale which include the discounts, taxes and capital cost increased over the years a little fluctuations is observed in 2018

```python
plt.figure(figsize=(15,8))
sb.lineplot(data=brs_df, x='Year', y='Total Orders')
plt.title('Trend of Total Orders Over Year')
plt.xlabel('Year')
plt.ylabel('Total Orders')
plt.show()
```

The amount of orders were on the increase over the years

```python
plt.figure(figsize=(15,8))
sb.lineplot(data=brs_df, x='Year', y='Total Sales')
plt.title('Trend of Total Sales Over Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sb.lineplot(data=brs_df, x='Month', y='Total Sales')
plt.title('Trend of Total Sales Over Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
